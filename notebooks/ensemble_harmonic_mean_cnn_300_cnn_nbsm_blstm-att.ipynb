```python
import pandas as pd
from glob import glob
import numpy as np
from scipy.stats import hmean
```

```python
subs = np.stack([pd.read_csv(sub).set_index('id').values for sub in glob('../submissions/*.csv')])
```

```python
preds = pd.read_csv('../data/sample_submission.csv')
```

```python
preds.iloc[:, 1:] = hmean(subs+1e-40, axis=0)
```

```python
preds.head()
```

```python
preds.to_csv('../submissions/harmonic_mean.csv', index=False)
```
