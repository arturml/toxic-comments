```python
import pandas as pd
import numpy as np
```

Using our name convetion it will take the features from ../data/features/features_model_name.csv validate a model and predict the test set with features from ../submission/model_name.csv

```python
models = ['naive_bayes', 'nblogistic', 'fasttext', 'GRU_pooling_att']
```

```python
features_path = '../data/features/'
submission_path = '../submissions/'
```

```python
features_files = [features_path + 'features_' + model + '.csv' for model in models]
submissions_files = [submission_path + model + '.csv' for model in models]
```

```python
target_df = pd.read_csv('../data/train.csv', index_col='id').drop('comment_text', axis=1)
```

```python
dfs = []
for i, f in enumerate(features_files):
    print(f)
    df = pd.read_csv(f, index_col='id')
    df.columns = df.columns + '_' + models[i]
    dfs.append(df)

features_df = pd.concat(dfs, axis=1).drop('Unnamed: 0_nblogistic', axis=1)
```

```python
features_df.head()
```

```python
dfs = []
for i, f in enumerate(submissions_files):
    print(f)
    df = pd.read_csv(f, index_col='id')
    df.columns = df.columns + '_' + models[i]
    dfs.append(df)

submissions_df = pd.concat(dfs, axis=1)
```

```python
submissions_df
```

# Parameter optimiztion for the ensemble classifier

```python
from sklearn.linear_model import LogisticRegressionCV
```

```python
probs = np.zeros(shape=(len(submissions_df), 6))
for i, col in enumerate(target_df.columns):
    print("Training ", col)
    y = target_df.loc[:, col].values
    ensemble_model_classifier = LogisticRegressionCV(Cs=30, n_jobs=10, refit=True)
    ensemble_model_classifier.fit(features_df, y)
    y_pred = ensemble_model_classifier.predict_proba(submissions_df)[:,1]
    probs[:, i] = y_pred
```

```python
sample_submission = pd.read_csv('../data/sample_submission.csv')
```

```python
sample_submission.iloc[:, 1:] = probs
```

```python
ensemble_file_name = '../submissions/ensemble_' + '_'.join(models) + '.csv'
ensemble_file_name
```

```python
sample_submission.to_csv(ensemble_file_name, index=False)
```
